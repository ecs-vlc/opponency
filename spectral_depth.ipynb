# Spectral opponency as a function of ventral depth
This notebook plots the distribution of spectrally opponent, non-opponent and unresponsive cells in different layers of our model as a function of ventral depth. It corresponds to Figure 7 in the Appendix.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from statistics import load

frame = pd.read_pickle('logs/devalois.pd')

plt.figure(figsize=(3,2.5))

opps = load.spectral(frame, 'spectrally opponent', groupby='d_vvs')
retina2 = opps[opps['layer'] == 'retina_relu2']
ventral0 = opps[opps['layer'] == 'ventral_relu0']
ventral1 = opps[opps['layer'] == 'ventral_relu1']

# When labelling, vantral_relu0 is called Ventral_1 for consistency, ventral_1 --> Ventral_2
plt.errorbar(retina2['d_vvs'], retina2['mean_rel_amount'], yerr=retina2['std_rel_amount'], label='Retina2', linestyle=':')
plt.errorbar(ventral0['d_vvs'], ventral0['mean_rel_amount'], yerr=ventral0['std_rel_amount'], label='Ventral1', linestyle='--')
plt.errorbar(ventral1['d_vvs'], ventral1['mean_rel_amount'], yerr=ventral1['std_rel_amount'], label='Ventral2', linestyle='-.')
x1,x2,y1,y2 = plt.axis()
plt.axis([x1,x2,y1,y2])

plt.xlabel('Ventral Depth')
plt.ylabel('Percentage \nSpectrally Opponent')
plt.xlim(1, 4)
plt.ylim(0, 1)
plt.savefig('figures/spectral_depth_opponent.eps', bbox_inches='tight')

plt.figure(figsize=(3,2.5))

opps = load.spectral(frame, 'spectrally non-opponent', groupby='d_vvs')
retina2 = opps[opps['layer'] == 'retina_relu2']
ventral0 = opps[opps['layer'] == 'ventral_relu0']
ventral1 = opps[opps['layer'] == 'ventral_relu1']

plt.errorbar(retina2['d_vvs'], retina2['mean_rel_amount'], yerr=retina2['std_rel_amount'], label='Retina2', linestyle=':')
plt.errorbar(ventral0['d_vvs'], ventral0['mean_rel_amount'], yerr=ventral0['std_rel_amount'], label='Ventral1', linestyle='--')
plt.errorbar(ventral1['d_vvs'], ventral1['mean_rel_amount'], yerr=ventral1['std_rel_amount'], label='Ventral2', linestyle='-.')
plt.axis([x1,x2,y1,y2])

plt.xlabel('Ventral Depth')
plt.ylabel('Percentage \nSpectrally Non-opponent')
plt.xlim(1, 4)
plt.ylim(0, 1)
plt.savefig('figures/spectral_depth_nonopponent.eps', bbox_inches='tight')

plt.figure(figsize=(3,2.5))

opps = load.spectral(frame, 'spectrally unresponsive', groupby='d_vvs')
retina2 = opps[opps['layer'] == 'retina_relu2']
ventral0 = opps[opps['layer'] == 'ventral_relu0']
ventral1 = opps[opps['layer'] == 'ventral_relu1']

plt.errorbar(retina2['d_vvs'], retina2['mean_rel_amount'], yerr=retina2['std_rel_amount'], label='Retina2', linestyle=':')
plt.errorbar(ventral0['d_vvs'], ventral0['mean_rel_amount'], yerr=ventral0['std_rel_amount'], label='Ventral1', linestyle='--')
plt.errorbar(ventral1['d_vvs'], ventral1['mean_rel_amount'], yerr=ventral1['std_rel_amount'], label='Ventral2', linestyle='-.')
plt.axis([x1,x2,y1,y2])

plt.legend()
plt.xlabel('Ventral Depth')
plt.ylabel('Percentage \nSpectrally Unresponsive')
plt.xlim(1, 4)
plt.ylim(0, 1)
plt.savefig('figures/spectral_depth_unresponsive.eps', bbox_inches='tight')
```
