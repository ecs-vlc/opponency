## Generate Plots

```python
import pandas as pd
import numpy as np

drifting = pd.read_csv('zhao_drifting.csv', names=['response', 'angle'], header=None)
drifting['angle'] = (drifting['angle'] * np.pi) / 180
drifting = drifting.sort_values(by='angle')
flashing = pd.read_csv('zhao_flashing.csv', names=['response', 'angle'], header=None)
flashing['angle'] = (flashing['angle'] * np.pi) / 180
flashing = flashing.sort_values(by='angle')

%matplotlib inline
import matplotlib.pyplot as plt

plt.polar(drifting['angle'], drifting['response'])
plt.gca().set_yticklabels([])
plt.gca().spines['polar'].set_visible(False)
plt.savefig("figures/zhao_drifting.pdf", bbox_inches='tight')

plt.figure()
plt.polar(flashing['angle'], flashing['response'])
plt.gca().set_yticklabels([])
plt.gca().spines['polar'].set_visible(False)
plt.savefig("figures/zhao_flashing.pdf", bbox_inches='tight')
```
