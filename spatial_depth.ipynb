```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from statistics import load

frame = pd.read_pickle('logs/spatial.pd')

plt.figure(figsize=(3,2.5))
opps = load.spatial(frame, 'spatially opponent', groupby='d_vvs')
retina2 = opps[opps['layer'] == 'retina_relu2']
ventral0 = opps[opps['layer'] == 'ventral_relu0']
ventral1 = opps[opps['layer'] == 'ventral_relu1']

plt.errorbar(retina2['d_vvs'], retina2['mean_rel_amount'], yerr=retina2['std_rel_amount'], label='Retina2', linestyle=':')
plt.errorbar(ventral0['d_vvs'], ventral0['mean_rel_amount'], yerr=ventral0['std_rel_amount'], label='Ventral1', linestyle='--')
plt.errorbar(ventral1['d_vvs'], ventral1['mean_rel_amount'], yerr=ventral1['std_rel_amount'], label='Ventral2', linestyle='-.')
x1,x2,y1,y2 = plt.axis()

plt.xlabel('Ventral Depth')
plt.ylabel('Percentage \nSpatially Opponent')
plt.xlim(1, 4)
plt.ylim(0, 1)
plt.savefig('figures/spatial_depth_opponent.eps', bbox_inches='tight')

plt.figure(figsize=(3,2.5))
opps = load.spatial(frame, 'spatially non-opponent', groupby='d_vvs')
retina2 = opps[opps['layer'] == 'retina_relu2']
ventral0 = opps[opps['layer'] == 'ventral_relu0']
ventral1 = opps[opps['layer'] == 'ventral_relu1']

plt.errorbar(retina2['d_vvs'], retina2['mean_rel_amount'], yerr=retina2['std_rel_amount'], label='Retina2', linestyle=':')
plt.errorbar(ventral0['d_vvs'], ventral0['mean_rel_amount'], yerr=ventral0['std_rel_amount'], label='Ventral1', linestyle='--')
plt.errorbar(ventral1['d_vvs'], ventral1['mean_rel_amount'], yerr=ventral1['std_rel_amount'], label='Ventral2', linestyle='-.')

plt.xlabel('Ventral Depth')
plt.ylabel('Percentage \nSpatially Non-opponent')
plt.xlim(1, 4)
plt.ylim(0, 1)
plt.savefig('figures/spatial_depth_nonopponent.eps', bbox_inches='tight')


plt.figure(figsize=(3,2.5))
opps = load.spatial(frame, 'spatially unresponsive', groupby='d_vvs')
retina2 = opps[opps['layer'] == 'retina_relu2']
ventral0 = opps[opps['layer'] == 'ventral_relu0']
ventral1 = opps[opps['layer'] == 'ventral_relu1']

plt.errorbar(retina2['d_vvs'], retina2['mean_rel_amount'], yerr=retina2['std_rel_amount'], label='Retina2', linestyle=':')
plt.errorbar(ventral0['d_vvs'], ventral0['mean_rel_amount'], yerr=ventral0['std_rel_amount'], label='Ventral1', linestyle='--')
plt.errorbar(ventral1['d_vvs'], ventral1['mean_rel_amount'], yerr=ventral1['std_rel_amount'], label='Ventral2', linestyle='-.')

plt.legend()
plt.xlabel('Ventral Depth')
plt.ylabel('Percentage \nSpatially Unresponsive')
plt.xlim(1, 4)
plt.ylim(0, 1)
plt.savefig('figures/spatial_depth_unresponsive.eps', bbox_inches='tight')
```
